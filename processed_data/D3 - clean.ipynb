```python
import pandas as pd 

df = pd.read_table(r'C:\Users\Francesca\Desktop\D3 - persons held (cleaned and dropped).tsv', sep='\t', header=0)
df
```

```python
to_sep = pd.DataFrame(df.iloc[:,0])
to_sep
```

```python
to_sep[['freq','indic_cr','unit','geo']] = to_sep["freq,indic_cr,unit,geo\TIME_PERIOD"].str.split(",", expand = True)
to_sep
```

```python
df = to_sep.merge(df, left_index=True, right_index=True)
df = df.drop(columns=["freq,indic_cr,unit,geo\TIME_PERIOD_x", "freq,indic_cr,unit,geo\TIME_PERIOD_y"])
df
```

```python
df.reset_index().T.drop_duplicates().T
```

```python
cleandf = df.drop(columns=['freq_x', 'freq_y', 'indic_cr_y', 'indic_cr_x', "unit_y", "geo_y"]) 

cleandf
```

```python
cleandf.rename(
    columns={"unit_x":"Unit", "geo_x": "Geo"}, 
    inplace=True)
cleandf
```

```python
cleandf.to_csv(r'C:\Users\Francesca\Desktop\cleandf.tsv')
```
