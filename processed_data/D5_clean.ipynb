```python
import pandas as pd 

df = pd.read_table('../raw_data/D5_CPI_from_eurostat.tsv', sep='\t', header=0)
df
```

```python
to_sep = pd.DataFrame(df.iloc[:,0])
to_sep
```

```python
to_sep[['freq', 'unit','geo']] = to_sep["freq,unit,geo\TIME_PERIOD"].str.split(",", expand = True)
to_sep
```

```python
df = to_sep.merge(df, left_index=True, right_index=True)
df = df.drop(columns=["freq,unit,geo\TIME_PERIOD_x", "freq,unit,geo\TIME_PERIOD_y"])
df
```

```python
cleandf = df.drop(columns=['freq'])

cleandf
```

```python
cleandf.to_csv('./clean_datasets/D5_clean.csv', index=False)
```
